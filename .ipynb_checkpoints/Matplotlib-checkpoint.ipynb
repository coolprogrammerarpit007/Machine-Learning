```python
# Basics of Matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("Number Growth")
plt.xlabel("X - AXIS")
plt.ylabel("Y - AXIS")
```

```python
# subplots in matplotlib -> it means making multiple plots together

plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
# Object Orieted Way

# firtly create an object

fig = plt.figure()  # this is a canvas obbject
axis1 = fig.add_axes([0.1,0.1,1,1])
axis2 = fig.add_axes([0.5,0.3,0.3,0.3])

axis1.plot(x,y)
axis2.plot(x,y)
```

```python
# Types of Plots

plt.scatter(x,y)
```

```python
from random import sample

data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x
           )
```

```python
# saving plots

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basic Plot.png")
```
