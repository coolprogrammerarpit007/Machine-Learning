```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Countries.csv')
```

```python
df.head()
```

```python
# data preprocessing

df.shape
```

```python
df.info()
```

```python
# which country has the hoghest population

df[df['population'] == df['population'].max()][['country','population']]
```

```python
# capital of the country of highest population
df[df['population'] == df['population'].max()][['country','population','capital_city']]
```

```python
# which country has the least population

df[df['population'] == df['population'].min()][['country','population','capital_city']]
```

```python
# top 5 countries with highest democratic score


df.sort_values(by='democracy_score',ascending=False,inplace=True)
```

```python
df['country'].head()
```

```python
# how many total regions there

df['region'].value_counts().count()
```

```python
# how many countries in the Eastern Europe Region

df['region'].value_counts()['Eastern Europe']
```

```python
# political leader of the second highest country

df[df['population'] == df['population'].nlargest(2).iloc[1]]
```

```python
# how many countries where political leader are unknown

df.loc[df['political_leader'].isna()]['country'].count()
```

```python
# how many countries there with republic in there full name
count = 0
def counting_countries(country):
    global count
    if  'republic' in country.lower():
        count += 1

    return country

df['country_long'].apply(counting_countries)
print(count)
```
