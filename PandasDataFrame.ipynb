```python
# Finding Missing Data
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}


df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# to check if in any column any Nan value exist

df.isna().any()
```

```python
# Removing Missing Data

# it works on the basis of rows
df.dropna()  # will remove row of null values
```

```python
df
```

```python
# Thrushold
df.dropna(thresh=3) # will drop only rows which has less than 3 Non Nan values



```

```python
# Filling the missing data

df
```

```python
df.fillna(0) # filling with zeros
```

```python
# fillling with different values for different ccolumns

values = {'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
